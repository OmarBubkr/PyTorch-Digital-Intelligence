```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

**Synthesizing Data**

```python
X = torch.randn(100, 1)*10
y = X + torch.randn(100, 1)
plt.plot(X.numpy(), y.numpy(), 'o')
plt.xlabel('x')
plt.ylabel('y')
```

**Constructing Linear Digital  Reasoning**

```python
class LR(nn.Module):
  def __init__(self, in_size, out_size):
    super().__init__()
    self.linear = nn.Linear(in_size, out_size)

  def forward(self, x):
    return self.linear(x)
```

```python
model = LR(1, 1)

[w, b] = model.parameters()
def get_params():
  return (w[0][0].item(), b[0].item())
```

**Inferring Reason**

```python
def plot_infer(title):
  plt.title = title  #
  w,b = get_params()
  x1 = np.array([-30, 30])
  y1 = w * x1 + b

  plt.plot(x1, y1, 'r')
  plt.scatter(X, y)
  plt.show()
```

```python
plot_infer('Digital Linear Reason')
```

**Training & Optimizing Linear Digital Brain**

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
epochs = 100
losses = []
for i in range(epochs):
  y_pred = model.forward(X)
  loss = criterion(y_pred, y)
  print("epoch:", i, "loss:", loss.item())

  losses.append(loss.item())
  optimizer.zero_grad() #
  loss.backward()
  optimizer.step()
```

```python
plt.plot(range(epochs), losses)
# plt.yscale("log")   # set y-axis to logarithmic
plt.ylabel('Loss')
plt.xlabel('epoch')
```

```python
plot_infer("Trained Model")
```
